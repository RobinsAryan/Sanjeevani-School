##1. Subtract the mean of each variable
To centre the dataset around the origin, subtract the mean of each variable from the total. This turns out to be a very useful step in the covariance matrix calculation process.

```python
X_meaned = X - np.mean(X , axis = 0)
```

##3. Calculate the covariance matrix of the centered data.
Determine the mean-centered data's covariance matrix. You can read this incredibly educational Wikipedia page to learn more about the covariance matrix.

The covariance matrix, which represents the covariance of the elements with one another, is a square matrix. An element's covariance with itself is equal to only its variance.

For this reason, the variance of each member makes up the diagonal elements of a covariance matrix.

```python
cov_mat = np.cov(X_meaned , rowvar = False)
```

select number of features we want data to be presented in


```python
num_factors = 6
```

##4. Compute the eigenvalues and eigenvectors of the covariance matrix.

Proceed to calculate the Eigenvalues and Eigenvectors of the Covariance matrix that has been computed. We obtain the orthogonal eigenvectors of the covariance matrix, where each vector corresponds to a principal axis.

A larger variability is correlated with a higher eigenvalue. As a result, the major axis that has the greater Eigenvalue will be one that represents more data variability.

When two vectors are mutually perpendicular, they are said to be orthogonal. Before we understand the notion and idea underlying eigenvalues and vectors, they appear to be very dangerous.

```python
eigen_values , eigen_vectors = np.linalg.eigh(cov_mat)
```

##Sort the eigenvalues and their corresponding eigenvectors in descending order.

Together with their matching Eigenvector, arrange the Eigenvalues in decreasing order.

Recall that each column in the Eigen vector-matrix represents a principal component; hence, the principal components will naturally be arranged in descending order of variability if the columns are arranged in descending order of their Eigenvalue.

Therefore, a main component that represents the most variability will be the first column in our rearranged Eigen vector-matrix.

```python
sorted_index = np.argsort(eigen_values)[::-1]
sorted_eigenvalue = eigen_values[sorted_index]
sorted_eigenvectors = eigen_vectors[:,sorted_index]
```

 ## Choose the top k eigenvectors,

 Select a subset from the rearranged Eigenvalue matrix as per our need i.e. number_comp = 6. This means we selected the first 6 principal components.

```python
eigenvector_subset = sorted_eigenvectors[:,0:num_factors]
```

## Project the original data onto the selected eigenvectors to obtain the reduced-dimensional data.

Lastly, convert the data by taking the dot product of the mean-centered data and the transpose of the Eigenvector subset. We can obtain the data reduced to lower dimensions from higher dimensions by transposing the dot product result.

```python
X_reduced = np.dot(eigenvector_subset.transpose() , X_meaned.transpose() ).transpose()
X_reduced
```

##Problem Formulation:

User-Item Matrix: In recommendation systems, the dataset is often represented as a user-item matrix, where rows correspond to users, columns correspond to items, and the entries represent user ratings or interactions with items. This matrix is often sparse because users interact with only a subset of items.

##Initialization:
Matrix factorization involves approximating the original matrix R by the product of two lower-rank matrices U and V.

##Objective Function:

The objective is to minimize the difference between the observed ratings in the original matrix R and the product of the factorized matrices U and V. This is often done by minimizing a cost function that includes a regularization term to prevent overfitting.

##Optimization:

Optimization techniques, such as gradient descent, stochastic gradient descent, or alternating least squares, are used to update the elements of U and V to minimize the objective function iteratively. This involves adjusting the values in U and V  to reduce the error between the predicted and actual ratings.

##Convergence:

The optimization process continues until a convergence criterion is met, such as a certain number of iterations or a small change in the cost function. Convergence ensures that the factorized matrices provide a good approximation of the original matrix.
##Prediction:

Once the factorization is complete, predictions for missing or unknown entries in the original matrix can be obtained by multiplying the corresponding elements in U and V.

```python

import numpy
def matrix_factorization(R, P, Q, K, steps=5000, alpha=0.0002, beta=0.02):

    Q = Q.T
    for step in range(steps):
        for i in range(len(R)):
            for j in range(len(R[i])):
                if R[i][j] > 0:
                    eij = R[i][j] - numpy.dot(P[i,:],Q[:,j])
                    for k in range(K):
                        P[i][k] = P[i][k] + alpha * (2 * eij * Q[k][j] - beta * P[i][k])
                        Q[k][j] = Q[k][j] + alpha * (2 * eij * P[i][k] - beta * Q[k][j])
        eR = numpy.dot(P,Q)
        e = 0
        for i in range(len(R)):
            for j in range(len(R[i])):
                if R[i][j] > 0:
                    e = e + pow(R[i][j] - numpy.dot(P[i,:],Q[:,j]), 2)
                    for k in range(K):
                        e = e + (beta/2) * (pow(P[i][k],2) + pow(Q[k][j],2))
        if e < 0.001:
            break
    return P, Q.T
```

##comparision

```python
from sklearn.decomposition import PCA, TruncatedSVD
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
import time

# Perform PCA
pca = PCA(n_components=num_factors)
start_time = time.time()
X_pca = pca.fit_transform(X_meaned)
pca_elapsed_time = time.time() - start_time
explained_variance_pca = np.sum(pca.explained_variance_ratio_)

# Perform Matrix Factorization (Truncated SVD)
svd = TruncatedSVD(n_components=num_factors)
start_time = time.time()
X_svd = svd.fit_transform(X_meaned)
svd_elapsed_time = time.time() - start_time
explained_variance_svd = np.sum(svd.explained_variance_ratio_)


# Calculate Reconstruction Error for PCA and SVD
data_reconstructed_pca = pca.inverse_transform(X_pca)
reconstruction_error_pca = mean_squared_error(X_meaned, data_reconstructed_pca)

data_reconstructed_svd = svd.inverse_transform(X_svd)
reconstruction_error_svd = mean_squared_error(X_meaned, data_reconstructed_svd)

# Print and compare the results
print("PCA:")
print("Explained Variance:", explained_variance_pca)
print("Reconstruction Error:", reconstruction_error_pca)
print("Elapsed Time:", pca_elapsed_time)

print("\nSVD:")
print("Explained Variance:", explained_variance_svd)
print("Reconstruction Error:", reconstruction_error_svd)
print("Elapsed Time:", svd_elapsed_time)
```
